```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Din 2014"
```

```python
df = pd.read_csv('./total_st_students-counselors.csv')
```

```python
df['Result'] = df['TOTSTU']/df['TOTCOU']
ratio = df.sort_values('Result', ascending=False)
ratio
```

```python
ax = ratio.sort_values(
    'Result', ascending=False
).head(5).plot(y="Result", kind="bar", color='#0D3C73', figsize=(10,7), rot=0, fontsize=12, legend=False
)
ax.set_color_cycle(['#0D3C73', '#329CAC', '#47BEF8', '#F95100', '#F7A12D'])
for p in ax.patches: ax.annotate(
    np.around(p.get_height(), decimals=0
            ), (p.get_x()+p.get_width()/2., p.get_height()), 
    ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=12)
plt.xlabel(
    "Source: National Center for Education Statistics", fontsize=12, color='#AB9F8F'
)
ax.xaxis.set_label_coords(.23, -0.1)
plt.suptitle('States with the highest counselor-to-student ratio for 2015 - 2016', y=.93, fontsize=16)
plt.xticks(np.arange(5), ('Arizona', 'Michigan', 'California', 'Minnesota', 'Utah'), fontsize=14)
plt.style.use('seaborn-darkgrid')

plt.savefig('./ratio.svg', bbox_inches = 'tight', pad_inches = 0)
plt.show()
```
