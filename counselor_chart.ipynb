```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./total_st_students-counselors.csv')
```

```python
df['Result'] = df['TOTSTU']/df['TOTCOU']
ratio = df.sort_values('Result', ascending=False)
ratio
```

```python
ax = ratio.sort_values(
    'Result', ascending=False
).head(5).plot(y="Result", kind="bar", figsize=(10,7),
               color='r', rot=0, fontsize=12, legend=False
)
for p in ax.patches: ax.annotate(
    np.round(p.get_height(),decimals=1
            ), (p.get_x()+p.get_width()/2., p.get_height()), 
    ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=12)
plt.xlabel(
    "Source: National Center for Education Statistics", fontsize=12
)
ax.xaxis.set_label_coords(.7, -0.1)
plt.suptitle('Counselor to student ratio for 2015 - 2016', fontsize=16)
plt.xticks(np.arange(5), ('Arizona', 'Michigan', 'California', 'Minnesota', 'Utah'), fontsize=14)
plt.style.use('seaborn-darkgrid')
plt.savefig('./ratio.svg', bbbox_inches='tight')
plt.show()
```
